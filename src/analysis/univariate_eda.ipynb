## Task 2: Perform data visualization to gain insights into the dataset (Univ)

### Understanding stock code/Desciption


```python
n = 20
most_popular_prods = df['Description'].value_counts().head(n)

plt.figure(figsize=(12, 6))
most_popular_prods.plot(kind='barh', color='skyblue')
plt.gca().invert_yaxis()  # Puts the highest selling product at the top
plt.xlabel('Number of Orders / Frequency')
plt.ylabel('Product Description')
plt.title(f'Top {n} Most Popular Products')
plt.tight_layout()
plt.show()
```

### Understanding stock invoices


```python
invoices = df['InvoiceNo'].value_counts()
invoices
```

```python
plt.figure(figsize=(12, 6))
invoices.head(n).plot(kind='barh', color='skyblue')
plt.gca().invert_yaxis()  # Puts the highest selling product at the top
plt.xlabel('Number products in the invoice')
plt.ylabel('Invoice number')
plt.title(f'Top {n} Biggest invoices')
plt.tight_layout()
plt.show()
```

```python
invoices = invoices.reset_index()

vis.plot_distribution(df=invoices, y='count', title='Invoice size', plt_type='both')
```

```python
display(invoices['count'].describe())
```

```python
display(invoices['count'].describe(percentiles=[i/10 for i in range(1, 10)]))
```

```python
invoices_no_outlier = invoices[invoices['count'] < 750]

vis.plot_distribution(df=invoices_no_outlier, y='count', title='Invoice size less than 750 distribution', plt_type='both')
```

```python
display(invoices_no_outlier['count'].describe(percentiles=[i/10 for i in range(1, 10)]))
```

#### Invoice size

Invoice size is highly right-skewed. The median invoice has 12 product lines, while the mean is about 22, indicating that a small number of very large invoices pull the average upward. 90% of invoices have 48 or fewer lines, but the maximum invoice has 1,114 lines.

This suggests that most invoices are relatively small, while a small group of invoices may represent bulk orders, operational records, or unusual transactions. These large invoices should be reviewed separately before using invoice size as a typical customer behavior metric.

### Understanding quantity

```python
vis.plot_distribution(df=df, y='Quantity', title='Quantity distribution', plt_type='both')
```

```python
df_quantity_no_outlier = df[abs(df["Quantity"]) < 10000]
vis.plot_distribution(df=df_quantity_no_outlier, y='Quantity', title='Quantity distribution less than 10000', plt_type='both')
```

```python
display(df_quantity_no_outlier['Quantity'].describe(percentiles=[i/10 for i in range(1, 10)]))
```

#### Quantity

Quantity contains extreme positive and negative values. The presence of negative quantities suggests returns, cancellations, or corrections rather than standard purchases. Because of this, analyzing all quantities together can mix different business processes.

For sales analysis, positive and negative quantities should likely be separated: positive quantities for purchases, negative quantities for returns/cancellations.

### Understanding UnitPrice

```python
vis.plot_distribution(df=df, y='UnitPrice', title='Unit Price distribution', plt_type='both')
```

```python
df_price_no_outlier = df[abs(df["UnitPrice"]) < 10000]

vis.plot_distribution(df=df_price_no_outlier, y='UnitPrice', title='Unit Price less than 10000 distribution', plt_type='both')
```

```python
display(df_price_no_outlier['UnitPrice'].describe(percentiles=[i/10 for i in range(1, 10)]))
```

#### UnitPrice

UnitPrice is strongly right-skewed. Most products have low unit prices: the median is around 2.10 and 90% of values are below 7.95 after removing the largest extreme values. However, there are very large prices that may correspond to special items, manual adjustments, postage, debt adjustments, or data quality issues.

Because UnitPrice alone does not represent sales impact, the next step should be to combine it with Quantity to analyze revenue.

### Understanding CustomerID

```python
customers_with_most_requests = df['CustomerID'].value_counts().head(n)

plt.figure(figsize=(12, 6))
customers_with_most_requests.head(n).plot(kind='barh', color='skyblue')
plt.gca().invert_yaxis()  # Puts the highest selling product at the top
plt.xlabel('Number products by Customer')
plt.ylabel('Invoice requests')
plt.title(f'Top {n} Customers')
plt.tight_layout()
plt.show()
```

```python
customers_with_most_requests_ids = customers_with_most_requests[customers_with_most_requests.index != 'missing']

plt.figure(figsize=(12, 6))
customers_with_most_requests_ids.plot(kind='barh', color='skyblue')
plt.gca().invert_yaxis()  # Puts the highest selling product at the top
plt.xlabel('Number products in the invoice by Customer')
plt.ylabel('Invoice requests')
plt.title(f'Top {n} Customers')
plt.tight_layout()
plt.show()
```

#### CustomerID

CustomerID should be interpreted as an identifier, not as a numeric variable. Its distribution is useful for counting transaction lines per customer, but averages or percentiles of CustomerID do not have business meaning.

The large number of missing customer IDs limits customer-level analysis. Future customer analysis should either exclude missing customers or analyze them as a separate group.

### Understanding Country

```python
plt.figure(figsize=(12, 6))
df['Country'].value_counts().head(n).plot(kind='barh', color='skyblue')
plt.gca().invert_yaxis()  # Puts the highest selling product at the top
plt.xlabel('Number products by Country')
plt.ylabel('Invoice requests')
plt.title(f'Top {n} Countries with more invoices')
plt.tight_layout()
plt.show()
```

### Country

The dataset is heavily concentrated in the United Kingdom. This means global conclusions will mostly reflect UK behavior unless countries are analyzed separately or normalized by their transaction volume.

For cross-country comparison, raw row counts are not enough. Revenue, number of invoices, number of customers, and average order value would give a more meaningful comparison.
