```python
from pathlib import Path
import sys

def find_project_root(start: Path = Path.cwd()) -> Path:
    for path in [start, *start.parents]:
        if (path / "README.md").exists() and (path / "data").exists():
            return path
    raise FileNotFoundError("Project root not found")

BASE_DIR = find_project_root()
print(f'Working base set at {BASE_DIR}')

if str(BASE_DIR) not in sys.path:
    sys.path.append(str(BASE_DIR))

DATA_DIR = BASE_DIR / "data"
purchases_path = DATA_DIR / 'Online Retail purchases.xlsx'
retail_path = DATA_DIR / 'Online Retail Cleaned.xlsx'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
purchase_invoices = pd.read_excel(purchases_path)
display(purchase_invoices.head(5))
```

```python
purchase_invoices.sort_values("revenue", ascending=False).head(10)
```

### Daily sales time series

```python
time_series = purchase_invoices.groupby('InvoiceDate')['revenue'].agg('sum').reset_index()

sns.lineplot(data=time_series, x="InvoiceDate", y="revenue")
plt.tight_layout()
plt.show()
```

## Limitations

This analysis is observational and descriptive. It does not establish causal relationships. Some records appear to represent returns, cancellations, discounts, or accounting adjustments, so purchases and non-purchase records should be analyzed separately. Customer-level conclusions are limited by the large number of missing CustomerID values. The dataset covers only one year of transactions, so seasonality findings should be interpreted cautiously.
